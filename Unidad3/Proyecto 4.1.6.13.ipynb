# Juego del gato

```python
from random import randrange

def DisplayBoard(board):
    print("+-------" * 3,"+",sep="")
    for row in range(3):
        print("|       " * 3,"|",sep="")
        for col in range(3):
            print("|   " + str(board[row][col]) + "   ",end="")
        print("|")
        print("|       " * 3,"|",sep="")
        print("+-------" * 3,"+",sep="")

def EnterMove(board):
    ok = False     # Suposición falsa, se necesita para entrar al ciclo
    while not ok:
        move = input("Ingresa tu movimiento: ")
        ok = len(move) == 1 and move >= '1' and move <= '9' # ¿Es válida la entrada del usuario?
        if not ok:
            print("Mal movimiento- repite tu entrada") # No lo es, intentalo de nuevo
            continue
        move = int(move) - 1      # Celdas de la 0 a la 8
        row = move // 3      # Filas de la celda
        col = move % 3      # Columnas de la celda
        sign = board[row][col]   # Verificar el cuadro seleccionado
        ok = sign not in ['O','X'] 
        if not ok:  # Está ocupado, ingrese nuevamente la entrada
            print("El campo ya está ocupado - repite tu entrada")
            continue
    board[row][col] = 'O' # Poner "O" en el cuadro seleccionado

def MakeListOfFreeFields(board):
    free = []    # La lista está inicialmente limpia
    for row in range(3): # Iterar a través de filas
        for col in range(3): # Iterar a través de columnas
            if board[row][col] not in ['O','X']: # ¿Está libre la celda?
                free.append((row,col)) # Sí lo está, agrega una tupa a la lista
    return free


def VictoryFor(board,sgn):
    if sgn == "X":  # ¿Buscamos una X?
        who = 'me'  # Sí, es del lado de la computadora
    elif sgn == "O": # ... ¿o para O?
        who = 'you'	# Sí, es nuestro lado
    else:
        who = None	# No deberiamos caer aquí
    cross1 = cross2 = True  # Para diagonales
    for rc in range(3):
        if board[rc][0] == sgn and board[rc][1] == sgn and board[rc][2] == sgn:	# Verificar fila rc
            return who
        if board[0][rc] == sgn and board[1][rc] == sgn and board[2][rc] == sgn: # Verificar columna rc
            return who
        if board[rc][rc] != sgn: # Verificar primera diagonal
            cross1 = False
        if board[2 - rc][2 - rc] != sgn: # Verificar segunda diagonal
            cross2 = False
    if cross1 or cross2:
        return who
    return None

def DrawMove(board):
    free = MakeListOfFreeFields(board) # Hacer una lista de campos vacios
    cnt = len(free)
    if cnt > 0: # Sí la lista no está vacia, escoger un lugar para 'X' y ponerlo
        this = randrange(cnt)
        row, col = free[this]
        board[row][col] = 'X'
        
board = [ [3 * j + i + 1 for i in range(3)] for j in range(3) ] # Hacer un tablero vacío
board[1][1] = 'X' # Poner inicialmente la 'X' en el centro
free = MakeListOfFreeFields(board)
humanturn = True # ¿De quién es el turno ahora?
while len(free):
    DisplayBoard(board)
    if humanturn:
        EnterMove(board)
        victor = VictoryFor(board,'O')
    else:
        DrawMove(board)
        victor = VictoryFor(board,'X')
    if victor != None:
        break
    humanturn = not humanturn
    free = MakeListOfFreeFields(board)

DisplayBoard(board)
if victor == 'you':
    print("¡Has ganado!")
elif victor == 'me':
    print("Gané")
else:
    print("¡Es un empate!")
```
